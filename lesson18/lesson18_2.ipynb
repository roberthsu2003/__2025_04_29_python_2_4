```python
import pandas as pd

df = pd.read_excel('各鄉鎮市區人口密度.xlsx',index_col=0)
df
```

### 一個DataFrame,如何將DataFrame做一個統計資料

```python
df1 = df[["年底人口數","土地面積","人口密度"]]

def statistic(s):
    return pd.Series([s.mean(),s.max(),s.min(),s.std(),s.median()],index=['平均','最高','最低','標準差','中間數'])

df1.apply(statistic,axis=0)
```
